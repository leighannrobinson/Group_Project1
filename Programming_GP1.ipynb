```python
import random

input("Welcome to Rock, Paper, Scissors! Please press Enter to start")

user_score=0
computer_score=0

while True:

    user_choice = input("Rock, Paper or Scissors?").lower()

    while user_choice != "rock" and user_choice != "paper" and user_choice != "scissors":

        user_choice = input("Invalid input, please try again: ").lower()

 

    random_num = random.randint(0,2)

    if random_num==0:

        computer_choice = "rock"

    elif random_num == 1:

        computer_choice = "paper"

    elif random_num == 2:

        computer_choice = "scissors"

   

    print()

    print("Your choice:", user_choice)

    print("Computer's choice:", computer_choice)

    print()

 

    if user_choice== "rock":

        if computer_choice=="rock":

            print("It's a tie!")

        elif computer_choice=="paper":

            print("You Lost!")

            computer_score +=1

        elif computer_choice=="scissors":

            print("You Win!")

            user_score +=1

    elif user_choice== "paper":

        if computer_choice=="paper":

            print("It's a tie!")

        elif computer_choice=="scissors":

            print("You Lost!")

            computer_score +=1

        elif computer_choice=="rock":

            print("You Win!")

            user_score +=1

    elif user_choice== "scissors":

        if computer_choice=="scissors":

            print("It's a tie!")

        elif computer_choice=="rock":

            print("You Lost!")

            computer_score +=1

        elif computer_choice=="paper":

            print("You Win!")

            user_score +=1

           

    print()

    print("You have", user_score,"wins")

    print("The computer has", computer_score,"wins")

    print()

   

    repeat=input("Play again? (Y/N) ").lower()

    while repeat != "n" and repeat != "y":

        repeat = input("Invalid input, please try again: ").lower()

       

    if repeat == "n":

        break

       

    print("\n----------------------------------------\n")
```
